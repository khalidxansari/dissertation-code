```python
from google.cloud import vision
import os

def detect_labels(path, all_labels):
    """Detects labels in the file and adds them to the all_labels dictionary."""
    client = vision.ImageAnnotatorClient()

    with open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)
    response = client.label_detection(image=image)
    labels = response.label_annotations

    print(f'Labels for {os.path.basename(path)}:')
    label_descriptions = []
    for label in labels:
        print(label.description)
        label_descriptions.append(label.description)
    print()  # Add an empty line for better readability

    if response.error.message:
        raise Exception(f'{response.error.message}')

    # Add labels to the all_labels dictionary
    all_labels[os.path.basename(path)] = label_descriptions

def detect_labels_in_directory(directory_path):
    """Detects labels in all images in the directory and saves them to a single file."""
    all_labels = {}
    for filename in os.listdir(directory_path):
        if filename.lower().endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif')):
            image_path = os.path.join(directory_path, filename)
            detect_labels(image_path, all_labels)

    # Write all labels to a single text file
    labels_file_path = os.path.join(directory_path, 'people_vision_tags.txt')
    with open(labels_file_path, 'w') as label_file:
        for image, labels in all_labels.items():
            label_file.write(f'Labels for {image}:\n')
            label_file.write('\n'.join(labels))
            label_file.write('\n\n')

# Path to your directory containing images
directory_path = 'D:/Kings College/Dissertation/Revised Image Datasets/People'

# Detect labels in all images in the directory
detect_labels_in_directory(directory_path)
```
